### For printing the SERS strips

```python
#here, we calculate th time that it takes to print a strip of SERS substrate
#given
#three min for 50 by 50 with pillar distance of 400-600nm
#how many second per square um
#it took 3 min (180 seconds to print 50 by 50 microns)
rate = 2500/180
#dimensions of strip in micrometer
height = 7000
#the width of the strip in micrometer
width = 10
area = height*width

time_for_strip = area/rate/3600
print('it takes ' + str(time_for_strip) + 'h to print the strip')
```

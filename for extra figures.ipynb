## to get a gradient

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the size of the gradient image
width = 10
height = 100

# Generate a gradient from dark blue to light blue
gradient = np.linspace(0, 1, height)[:, np.newaxis]

# Create a 2D array representing the gradient
gradient = np.repeat(gradient, width, axis=1)

# Plot the gradient as an image
plt.imshow(gradient, cmap='Blues', aspect='auto')
plt.axis('off')  # Turn off axis
plt.show()
```
